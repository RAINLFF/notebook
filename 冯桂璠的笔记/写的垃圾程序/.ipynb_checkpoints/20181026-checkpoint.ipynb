```python
from matplotlib import pyplot as plt
from matplotlib import animation
import numpy as np
from registernumber import dict as dit1
from topic_create import dict as dit2
fig = plt.figure()

ax = plt.axes()

dict = {}
for key,value in dit1.items():
    list = []
    list.append(value)
    list.append(dit2[key])
    dict[key] = list
line1, = ax.plot(dict[key],dict[key][0])
line2, = ax.plot(dict[key],dict[key][1])
def animate(i):
    line1.set_ydata(dict[2002+i][0])
    line2.set_ydata(dict[2002+i][1])
    return line1,line2

def init():
    line1.set_ydata(dict[2002][0])
    line2.set_ydata(dict[2002][1])
    return line1,line2

ani = animation.FuncAnimation(fig=fig,
                              func=animate,
                              frames=10,
                              init_func=init,
                              interval=1,
                              blit=False)

plt.show()
```

```python
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import numpy as np
engine = create_engine('')

sql = '''
        select last_login_time,register_time from user;
        '''
dict = {}
df = pd.read_sql_query(sql,engine)
for index,row in df.iterrows():
    try:
        across = row['last_login_time'].year-row['register_time'].year + 1
    except:
        continue
    if across <0:
        across = 0;
    if across not in dict:
        dict[across] = 1
    else:
        dict[across] += 1
plt.figure()
x_ticks = []
for i in range(0,18):
    x_ticks.append(i)

plt.xticks(x_ticks)
for x, y in dict.items():
    # ha: horizontal alignmentf
    # va: vertical alignment
    plt.text(x,y, '%s' % y, ha='center', va='bottom')

plt.bar(dict.keys(),dict.values(),facecolor='#9999ff', edgecolor='white')
plt.show()

#
```

```python
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import numpy as np
engine = create_engine('')

sql = '''
        select register_time from user;
        '''

df = pd.read_sql_query(sql,engine)
dict = {}

for item in df.register_time:
    if item.year not in dict:
        dict[item.year] = 1
    else:
        dict[item.year] += 1
plt.figure()

x_ticks = []
print(dict)
for i in range(2000,2018):
    x_ticks.append(i)

for x, y in dict.items():
    # ha: horizontal alignmentf
    # va: vertical alignment
    plt.text(x,y, '%s' % y, ha='center', va='bottom')

plt.xticks(x_ticks)
plt.ylim(0,15000)
plt.bar(dict.keys(),dict.values(),facecolor='#9999ff', edgecolor='white')
plt.show()

#
```

```python
package pratice_1;

public class QuickSort {
    public void sort(int n[]) {
        int left = 0;
        int right = n.length -1;
        quicksort(n,left,right);
    }
    static void quicksort(int n[],int left,int right){
        int mid;
        if(left<right) {
            mid = partition(n, left, right);
            quicksort(n, left, mid - 1);
            quicksort(n, mid + 1, right);
        }
    }
    static int partition(int n[],int left,int right){
        int i = n[left];
        while(left<right){
            while(left<right&& i<=n[right])
                right--;
            if(left<right)
                n[left++]=n[right];
            while(left<right&& i >=n[left])
                left++;
            if(left<right)
                n[right--]=n[left];
        }
        n[left] = i;
        return left;
    }
}
```
